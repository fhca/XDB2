## 1. Localizar y cargar a Pandas una tabla con las ciudades más grandes del mundo (100)

```python
# importa el módulo Pandas con el nombre "pd"
import pandas as pd
# importa el módulo Numpy con el nombre "np"
import numpy as np
# importa el módulo pyplot de matplotlib con el nombre "plt"
import matplotlib.pyplot as plt
# grafica dentro de la libreta (en vez de en una ventana externa)
%matplotlib inline

# Lee de una página web. Siempre lee de las <table> que encuentre y genera una lista de DataFrame
# Aquí extraemos la segunda (podemos listar las que saca sin el [1], y ver cual necesitaremos)
# header=0 dice que el primer renglón es de etiquetas para columnas.
ciudades = pd.read_html("https://en.wikipedia.org/wiki/World%27s_largest_cities", header=0)[1]
# extrae sólo dos columnas

ciudades = ciudades[['City', 'Urban area population[7]']]
ciudades.head()
```

```python
ciudades.columns
```

```python
# OPCIONAL
# diccionario para traducir los nombres de las columnas
traducción = {'City':'Ciudad', 'Urban area population[7]':'Población del área urbana'}
# traduce sus nombres
ct = ciudades.rename(columns=traducción) #opcional
ct
```

## 4. Generar una tabla de 0's y 1's (100x100) a partir de 10000 números seudoaleatorios decimales entre 0 y 1 con un 40% de 1's y un 60% de 0's

```python
# rand genera los números con distribución uniforme, indispensable para asegurar las proporciones pedidas
df = pd.DataFrame(np.random.rand(10000).reshape((100, 100)))
```

```python
df.head()
```

```python
# como los números están distribuidos uniformemente, el decidir para el valor de x equivale a 
# considerar la probabilidad de que aparezcan los valores con los porcentajes solicitados.
df.applymap(lambda x: 1 if x < .4 else 0)
```
